The state of a channel in a point in time is probablistic as a function of the energy required to shift from it's current state to the next. The figure below describes these energy states. Both the open state ($E_{O}$) and the closed ($E_{C}$) are stable states for the channel. Thus to change from one to the next, an energy barrier must be overcome ($\Delta E$). This means that the rate of channels switching from open to closed ($r_{OC}$) is greater than that of closed to open ($r_{CO}$) as the higher energy state of $E_{O}$ leads to less energy required to initiate a change of state. 

$\Delta E$ is formally $E_{O} - E_{C}$ but can be approximated as a function of average thermal noise, $k_{B}T/2$. These systems evolved under the conditions of thermal noise and interactions of molecules at a scale this minute are likely to be reliant on the phenomenon to drive functional change. Thus, the probability of a change in channel state is described in Boltzmann's distribution law, $\frac{P_{O}}{P_{C}} = e^{-\Delta E / k_{B} T}$. #### Energy States
Referring back to our earlier channel state probability function, $\frac{P_{O}}{P_{C}} = e^{-\Delta E / k_{B} T}$, we can eliminate $P_{C}$ using $P_{O}+P_{C}=1$, resulting in $P_{O} \big(\Delta E\big)  = \frac{1}{1+e^{\Delta E / k_{B} T}}$ . The following figure demonstrates the energy changes ($\Delta E$) between channel states as a function of a linear spring (the tip link) and internal energy configurations ($E^{*}$).

We can collate these energy states to form an expression of $\Delta E$. 

$$\Delta E = E_{O} - E_{C} = \Delta E^{*} + d^{2} - kxd$$

Substituting $z=kd$ for single channel gating force and $x_{ref} = \frac{\Delta E^{*} + d^{2}}{kd}$, we arrive at our function:

$$ \Delta E = - z \big(x-x_{ref}\big) $$

Thanks to the work of Howard, Hudspeth and Roberts (1988), we know that the gating force of a single channel ($z$) is $40fN$.

Now we can complete our open state expression, substituting $\Delta E$ to give $P_{O} \big(x\big)  = \frac{1}{1+e^{-z \big(x-x_{ref}\big) / k_{B} T}}$. Note that $P_{O} = P_{C}$ when $x=x_{ref}$, and the dynamic range of a hair cell is $ 0 \rightarrow Nkd^{2} $, where N is the total number of channels.

We are now in a position to model mechanotransduction in a hair cell. Channel states are samples from a Boltzmann distribution and thus the number of open channels is a sample of a Binomial distribution.

#### Side note: Molecular Computation
Computation occurs not only at the level of neuronal biology but also at level of the molecular. An example of such computation is the LAC operon, discovered by Joacon and Monod in 1961 and earning them a Nobel Prize in 1965. The structure of these group of genes allows for the prioritisation of glucose metabolism over lactose metabolism. Operon mechanisms are described in the figure below, from Sterling and Laughlin (2015).

For RNA to bind the LAC operon, two things need to occur: the CAP promotor protein must bind the promotoe region, and the repressor blocking the RNA binding site must dissociate. When glucose is present in the cell, levels of cAMP drop, decreasing the probability of activation of the cAMP-dependent protein CAP that acts as a promoter of the LAC operon. Thus, under presence of glucose, the operon is prevented from undergoing transcription, the protein for lactose metabolism is not produced and glucose metabolism is instead prioritised. This is shown in the bottom two states of the above figure. When glucose is not present in the cell, cAMP levels rise and CAP binds to the promotor region of the operon. However, transcription only occurs when lactose is present in the cell to bind the repressor protein that is present to prevent RNA binding in the absence of lactose, causing it to dissociate from the operon. Thus, the LAC operon for lactose metabolism is only transcribed in the presence of lactose and the absence of glucose, a process driven purely by molecular computation.

Processes such as these describe cells as statistical molecular machines (or self-organising thermodynamic structures) that manipulate probabilistic bindings, conformations, and affinities to compute on miniscule spatial and temporal scales, acquiring information to conserve energy and reduce effort. Nervous systems scale these machines to far greater dimensions, allowing the limits of diffusion (the primary driving force of molecular computation) to be overcome.

Hodgkin & Huxley famously devised a conduction-based model of the action potential in 1952, describing the kinetics and conductances of sodium and potassium channels within an action potential and how these related to the characteristic voltage changes of the membrane. These are summarised in the diagram below.

While we have discussed channel states as they relate to our hair model, we have taken a simplistic approach. Instead of a open-closed state binary, membrane channels operate through at least three individual conformations: a conducting active state, a non-conducting inactived state and a non-conducting deactivated state. These states characterise the refractory periods of the action potential. While the active state is self-explanatory, the inactivated channel is closed and cannot reopen until it has passed into the deactivated state. Thus, the deactivated channel is closed yet possesses the ability to activate. The dwell times that each channel spends in each state contributes to the conductance of the membrane and thus concerns itself with our model.

Gillespie's method of modelling channel states asserts that dwell times are exponentially distributed. The probability distribution of open intervals, $ f_{o}\big(\tau\big) $, is thus defined by $ r_{c}e^{-r_{c}\tau} $ where $r_{c}$ is the rate of channels switching to closed. Likewise, the probability distribution of open intervals, $ f_{c}\big(\tau\big) $, is defined by $ r_{o}e^{-r_{o}\tau} $ where $r_{o}$ is the rate of channels switching to open. To simulate the kinetics of a channel, all we have to do is run a simple algorithm of alternatively drawing a dwell time from each probability distribution. These values are exact in that they are simulated to machine precision, and extendable to multiple states.

#### Discrete-Time Ensemble Simulation

While Gillespie's method provides a functional approach to simulating channel state kinetics, the temporal resolution that it provides us for the channel kinetics of hair cells far surpasses the temporal resolution required for functional vestibular kinetics ie. the channel state probabilities reach equilibrium within a simulation time-step. Thus, channel states are independent trials at each simulation step, and transduction channel conductance can be simulated as a scaled binomial distribution: 

$$ G \sim g_{MET} B \big(N,p_{o}\big(x\big)\big) $$

where $g_{MET}$ is single-channel open state conductance, $N$ is the number of transduction channels and $p_{o}\big(x\big)$ is the open state probability as a function of gating spring extension $x$.
