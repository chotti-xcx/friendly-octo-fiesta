In order to add mechanotransduction mechanisms to our earlier hair cell model, we will describe the ion channel kinetics as a two state model: either the channel is active and open or inactive and closed. As shown above, the open state of the channel will allow influx of calcium and potassium ions. Due to concentration gradients over the cell membrane, potassium ions tend to enter the cell. With a conductance of 200 picosiemens, a single active mechanoreceptive channel opens a current of one picoampere for about  100 microseconds before switching to it's inactive state . Factors influencing channel activation and inactivation are included as sensory stimuli physically displacing the mechanoreceptive channel and adaptation or bias of the cell. Also included in the figure is the voltage change of 50 microvolts across the cell membrane, resulting in an extreme electric field in the magnitude of $10^{7} V m^{-1}$.  

The state of a channel in a point in time is probablistic as a function of the energy required to shift from it's current state to the next. The figure below describes these energy states. Both the open state ($E_{O}$) and the closed ($E_{C}$) are stable states for the channel. Thus to change from one to the next, an energy barrier must be overcome ($\Delta E$). This means that the rate of channels switching from open to closed ($r_{OC}$) is greater than that of closed to open ($r_{CO}$) as the higher energy state of $E_{O}$ leads to less energy required to initiate a change of state. 

$\Delta E$ is formally $E_{O} - E_{C}$ but can be approximated as a function of average thermal noise, $k_{B}T/2$. These systems evolved under the conditions of thermal noise and interactions of molecules at a scale this minute are likely to be reliant on the phenomenon to drive functional change. Thus, the probability of a change in channel state is described in Boltzmann's distribution law, $\frac{P_{O}}{P_{C}} = e^{-\Delta E / k_{B} T}$.
